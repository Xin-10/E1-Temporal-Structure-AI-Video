```python
import os
import pandas as pd
```

```python
root = r"C:\Users\lixin\Desktop\E1"
video_folder = os.path.join(root, "AI_videos")
prompts_csv = os.path.join(root, "AI_features", "Prompts.csv")
videos_csv = os.path.join(root, "AI_features", "Videos.csv")
```

```python
df_prompts = pd.read_csv(prompts_csv)
df_prompts = df_prompts.loc[:, ~df_prompts.columns.str.contains('^Unnamed')]

df_videos_meta = pd.read_csv(videos_csv)
df_videos_meta = df_videos_meta.loc[:, ~df_videos_meta.columns.str.contains('^Unnamed')]
```

```python
video_files = []
for file in os.listdir(video_folder):
    if file.lower().endswith((".mp4", ".mov")):
        video_files.append({"filename": file, "video_id": os.path.splitext(file)[0]})

df_folder = pd.DataFrame(video_files)
```

```python
df = df_videos_meta.merge(df_prompts, on="prompt_id", how="left")
df = df.merge(df_folder, on="video_id", how="left")
```

```python
df["filepath"] = df["filename"].apply(lambda x: os.path.join(video_folder, x))
```

```python
df = df.drop(columns=["file_path"])
```
